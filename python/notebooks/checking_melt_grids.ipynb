# Why arent the melt values created by square_melt_region using differnt melt_magnitudes exact multiples of each other?

```python
import xarray as xr
import hvplot.xarray
import numpy as np 
```

```python
results = xr.open_dataset('results_small_test.nc')
x_resolution = results.x[1].values - results.x[0].values
assert x_resolution == np.abs(results.y[1].values - results.y[0].values)
resolution = x_resolution
cell_area = resolution**2
```

```python
total_melt = (results['melt']*cell_area).sum(dim=['x','y'])
total_melt.name = 'total_melt'
total_melt.attrs['units'] = 'm3'
total_melt
total_melt.isel(melt_magnitude=1, x_center_of_melt=0)/
```
